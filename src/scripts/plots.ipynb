```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data from the table
number_of_nodes = [5, 10, 20, 100, 200]
no_cache_avg_time = [4.198177, 4.612834, 7.282058, 49.04187, 110.955029]
no_cache_avg_calls = [3.0165, 3.264, 3.9285, 5.08, 5.688]
cache_avg_time = [3.114864, 3.932333, 6.503199, 47.477797, 111.631045]
cache_avg_calls = [2.063, 2.719, 3.5195, 5.087, 5.708]

# Plot average time with specified colors
plt.figure(figsize=(10, 6))
plt.plot(number_of_nodes, no_cache_avg_time, marker='o', color='orange', label='No Cache - Avg Time (ms)')
plt.plot(number_of_nodes, cache_avg_time, marker='o', color='blue', label='Cache - Avg Time (ms)')
plt.xlabel('Number of Nodes')
plt.ylabel('Average Time (ms)')
plt.title('Average Time vs Number of Nodes')
plt.legend()
plt.grid(True)
plt.show()

# Plot average number of calls with specified colors
plt.figure(figsize=(10, 6))
plt.plot(number_of_nodes, no_cache_avg_calls, marker='o', color='orange', label='No Cache - Avg Number of Calls')
plt.plot(number_of_nodes, cache_avg_calls, marker='o', color='blue', label='Cache - Avg Number of Calls')
plt.xlabel('Number of Nodes')
plt.ylabel('Average Number of Calls')
plt.title('Average Number of Calls vs Number of Nodes')
plt.legend()
plt.grid(True)
plt.show()
```
